```python
from sklearn.tree import DecisionTreeClassifier
import csv
import numpy as np

filename = "buy_computer.csv"
with open(filename, 'r') as f:
    data = list(csv.reader(f))

data = np.array(data)
data
    
```

```python
X = data[1:, 1:5]
y = data[1:, 5]
```

```python
X
```

```python
y
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 4)
tree_clf.fit(X,y)
```

ValueError, so we will try a digitalized version of the csv

```python
filename = "buy_computer_digital.csv"
with open(filename, 'r') as f:
    data = list(csv.reader(f))

data = np.array(data)
data

X = data[1:, 1:5]
y = data[1:, 5]

data
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 4)
tree_clf.fit(X,y)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file = "buy_computer_tree.dot",
        feature_names=data[0, 1:5],
        class_names=data[0, 5],
        rounded=True,
        filled=True
)
```
